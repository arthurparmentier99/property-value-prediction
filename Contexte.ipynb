```python
percentage_missing = (df.isna().sum() / len(df)) * 100
print(percentage_missing)
```

```python
df.drop("Identifiant de document", axis=1, inplace=True)
df
```

```python
for colonne in df:
    if percentage_missing[colonne] > 80:
       df.drop(colonne, axis=1, inplace=True) 
print(df.shape)
```
